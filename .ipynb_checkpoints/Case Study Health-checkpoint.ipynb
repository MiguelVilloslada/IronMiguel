```python
list2=list(health1['gender'])
genderr=[]
for i in list2:
    if i == 'f' or i == 'female' or i == 'Female' or i=='F':
        genderr.append('female')
    elif i == 'm'or i == 'male' or i == 'Male' or i=='M':
        genderr.append('male')
    else:
        genderr.append('Unknown')
health1['gender'] = genderr
#Roger solution
```

```python
gender_list = []
for i in range(len(health1["gender"])):
    if health1["gender"][i] == "male" or health1["gender"][i] == "Male" or health1["gender"][i] == "M":
        gender_list.append("male")
    elif health1["gender"][i] == "female" or health1["gender"][i] == "Female" or health1["gender"][i] == "F":
        gender_list.append("female")
    else:
        gender_list.append("unknown")
health1["gender"] = gender_list
#Lucas Solution
```

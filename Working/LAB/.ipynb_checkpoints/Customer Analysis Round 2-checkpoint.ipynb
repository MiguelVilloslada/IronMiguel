```python
import pandas as pd
```

```python
mca = pd.read_csv("marketing_customer_analysis.csv")
```

```python
mca
```

```python
#Show the dataframe shape
mca.shape
```

```python
#Change uppercase to lower case 
columns = []
for i in range (len(mca.columns)):
    columns.append(mca.columns[i].lower())
```

```python
columns
```

```python
mca.columns = columns
```

```python
mca.columns
```

```python
#unnamed: 0 seems to be same as index so I am deleting it
mca = mca.drop(["unnamed: 0"],axis=1)
```

```python
mca.columns
```

```python
#rename column employment status
mca = mca.rename({'employmentstatus': 'employment status'}, axis=1)
```

```python
mca.columns
```

```python
#8 columns are numerical
mca.describe()
```

```python
#Check and deal with NaN values
check_nan_in_mca = mca.isnull().values.any()
print (check_nan_in_mca)
```

```python
mca.isnull().sum().sum()
```

```python
mca.info()
```

```python
mca["response"] =mca["response"].fillna("unknown")
```

```python
mca["months since last claim"] =mca["months since last claim"].fillna("unknown")
```

```python
mca["number of open complaints"] =mca["number of open complaints"].fillna("unknown")
```

```python
mca["vehicle class"] =mca["vehicle class"].fillna("unknown")
```

```python
mca["vehicle size"] =mca["vehicle size"].fillna("unknown")
```

```python
mca["vehicle type"] =mca["vehicle type"].fillna("unknown")
```

```python
mca.isnull().sum().sum()
```

```python
mca.head(10)
```

```python
#change object to date time (use a date time as a date time)
mca['effective to date']=pd.to_datetime(mca['effective to date'],errors='coerce')
```

```python
mca.info()
```

```python
#Extract the month in a separate column
mca['month']=pd.DatetimeIndex(mca['effective to date']).month
```

Show a plot of the distribution of the Total Claim Amount, broken down by response type. Try a boxplot and distribution plot, for each response type. Create a scatterplot between Total Claim Amount and Income. Play around with the settings of the scatterplot (markersize, alpha level, ...) and in doing so try to identify more features within the data just visually. You can also try different seaborn plots. 

```python
plt.scatter(x='total claim amount', y='income', s=5, c='red', data=mca);
plt.xlabel('total claim amount')
plt.ylabel('income')
plt.title('income v claim amount')
```
